```python
import os
import sys
import dlib
import cv2
import openface
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
def detect_and_align(file_path, person_name, id):
    predictor_model = os.path.join(os.getcwd(), "models","shape_predictor_68_face_landmarks.dat")

    
    face_detector = dlib.get_frontal_face_detector()
    face_pose_predictor = dlib.shape_predictor(predictor_model)
    face_aligner = openface.AlignDlib(predictor_model)

    file_name = file_path

    image = cv2.imread(file_name)

    
    detected_faces = face_detector(image, 1)

    print("Found {} faces in the image file {}".format(len(detected_faces), file_name))

    for i, face_rect in enumerate(detected_faces):

        print("- Face #{} found at Left: {} Top: {} Right: {} Bottom: {}".format(i, face_rect.left(), face_rect.top(), face_rect.right(), face_rect.bottom()))

        pose_landmarks = face_pose_predictor(image, face_rect)

        alignedFace = face_aligner.align(96, image, face_rect, landmarkIndices=openface.AlignDlib.OUTER_EYES_AND_NOSE)

        if(not os.path.exists(os.path.join(os.getcwd(), "aligned-images", person_name))):
            os.makedirs(os.path.join(os.getcwd(), "aligned-images", person_name))
        
        
        cv2.imwrite(os.path.join(os.getcwd(), "aligned-images", person_name, "{}.{}.jpg".format(id, i)), alignedFace)
```

```python
for root, dirs, files in os.walk("./train-images/modi-images"):
    for file in files:
        detect_and_align(os.path.join("./train-images/modi-images", file), "modi-images", int(file.split(".")[0]));
        
for root, dirs, files in os.walk("./train-images/kejriwal-images"):
    for file in files:
        detect_and_align(os.path.join("./train-images/kejriwal-images", file), "kejriwal-images", int(file.split(".")[0]));
```

```python
import csv

with open('embeddings.csv', 'wt') as csv_file:
    writer = csv.writer(csv_file, delimiter=',');
    net = openface.TorchNeuralNet()
    for root, dirs, files in os.walk("./aligned-images/modi-images"):
        for file in files:
            image = cv2.imread("./aligned-images/modi-images/"+file);
            embeddings = net.forward(image)
            row = embeddings.tolist()
            row.append("modi");
            row.append(file.split('.')[0])
            writer.writerow(row)
            
    for root, dirs, files in os.walk("./aligned-images/kejriwal-images"):
        for file in files:
            image = cv2.imread("./aligned-images/kejriwal-images/"+file);
            embeddings = net.forward(image)
            row = embeddings.tolist()
            row.append("kejriwal");
            row.append(file.split('.')[0])
            writer.writerow(row)
```

```python
import pandas as pd

dataset = pd.read_csv("embeddings.csv", header=None)
```

```python
dataset[128].replace(['modi', 'kejriwal'], [1,2], inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x = dataset.loc[:, :127].values
y = dataset.loc[:, 128].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.svm import SVC

clf = SVC(C=1, kernel='linear', probability=True)
clf.fit(x_train, y_train);

score = clf.score(x_test, y_test);
print('Linear svm score', score)
```

```python
from sklearn.svm import SVC

clf = SVC(C=1, kernel='rbf', probability=True, gamma=2)
clf.fit(x_train, y_train);

score = clf.score(x_test, y_test);
print('radial svm score', score)
```

## function to predict and draw

```python
def recog_face(file_path):
    img = cv2.imread(file_path);

    face_detector = dlib.get_frontal_face_detector()

    predictor_model = os.path.join(os.getcwd(), "models","shape_predictor_68_face_landmarks.dat")

    detected_faces = face_detector(img, 1)
    
    print("Found {} faces".format(len(detected_faces)))
    
    face_pose_predictor = dlib.shape_predictor(predictor_model)
    face_aligner = openface.AlignDlib(predictor_model)

    
    for i, face_rect in enumerate(detected_faces):

        pose_landmarks = face_pose_predictor(img, face_rect)

        alignedFace = face_aligner.align(96, img, face_rect, landmarkIndices=openface.AlignDlib.OUTER_EYES_AND_NOSE)

        net = openface.TorchNeuralNet()
        embeddings = net.forward(alignedFace)

        
        prediction = clf.predict_proba([embeddings]).ravel()
        
        
        if(prediction[0] > prediction[1]):
            name = 'modi'.format(max(prediction))
        else:
            name = 'kejriwal'.format(max(prediction))
            
        if(max(prediction) < 0.77):
            name = "someone else"

        cv2.rectangle(img,(face_rect.left(),face_rect.top()),(face_rect.right(),face_rect.bottom()),(0,255,0),2)
        
        cv2.putText(img, name, (face_rect.left(),face_rect.top()), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255), 2)
        cv2.putText(img, str(max(prediction)), (face_rect.left(),face_rect.bottom()), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255), 2)
        
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    
    print("====================================================================")
```

```python
for root, dirs, files in os.walk("./test-images"):
    for file in files:
        file = os.path.join(root,file);
        print(file);
        recog_face(file);
```
